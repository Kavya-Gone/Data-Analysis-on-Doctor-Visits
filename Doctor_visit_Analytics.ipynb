```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("DoctorVisits.csv")
df.head(5)
```

```python
df.info()
```

```python
df["income"].value_counts()
```

```python
df["gender"].value_counts()
```

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

```python
df.groupby(['gender','health']).mean()
```

```python
sns.heatmap(df.isnull(),cbar=True,linecolor='Red',cmap='GnBu')
```

```python
corr = df.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(df.income, df.visits,'o',markersize=2, color='brown')
plt.xlabel('income')
plt.ylabel('visits')
```

```python
label=['yes','no']
Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(3,4))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to low income")
plt.show()
Y=df[df['private']=='yes']
N=df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(3,4))
plt.pie(x,labels=label)
plt.title("% of people getting private health insurance ")
plt.show()
Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(4,3))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to old age,disability or veteran status")
plt.show()
```

```python
db=df.groupby('gender')['health'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['health'],color=["#7eb0d5" ,"#b2e061"])
plt.title("Overall Health Analysis")
plt.xlabel('Health')
plt.ylabel('Gender')
plt.show()
```
